Generate a dataset of 100 points 

```python
import numpy as np 
from random import randint
import matplotlib.pyplot as plt 
from numpy.random import rand
x = np.linspace(0,1,100)
```

Calculate the label and each x with the function


```python
y = np.sin(2*np.pi*x)
Noise=rand(100)-0.5
fx = y+Noise 
plt.figure(figsize=(7,5), constrained_layout=True)
plt.scatter(x, fx, color='red', label='Sin(2 Pi X )')
```

```python
t = [0.01, -0.23, 0.01, 0.01,0.01,0.01,0.01]

def poli(t,x):
    #eval = t[0]+t[1]*x+t[2]*((x)**2)+t[3]*((x)**3)
    eval = t[0]+t[1]*x+t[2]*((x)**2)+t[3]*((x)**3)+t[4]*((x)**4)+t[5]*((x)**5)+t[6]*((x)**6)
    return eval 
x2 = np.arange(0,1,0.01)
y2 = poli(t,x2)
plt.scatter(x,fx, color= 'red', label='Sin(2 pi x )')
plt.plot(x2, y2, 'b-')
```

```python
from IPython.core.display import clear_output
from IPython import display

alpha = 0.4
epoch=200
J = np.zeros(epoch)
for ep in range(epoch):
    for i in range(len(x)):
        t[0] = t[0] + alpha *(fx[i]-poli(t,x[i]))*(x[i]**0)
        t[1] = t[1] + alpha *(fx[i]-poli(t,x[i]))*(x[i]**1)
        t[2] = t[2] + alpha *(fx[i]-poli(t,x[i]))*(x[i]**2)
        t[3] = t[3] + alpha *(fx[i]-poli(t,x[i]))*(x[i]**3)
        t[4] = t[4] + alpha *(fx[i]-poli(t,x[i]))*(x[1]**4)
        t[5] = t[5] + alpha *(fx[i]-poli(t,x[i]))*(x[1]**5)
        t[6] = t[6] + alpha *(fx[i]-poli(t,x[i]))*(x[1]**6)
    y2 =poli(t,x)
    err =y2-y
    J_scalar = 0
    for i in range(len(x)):
        J_scalar=(1/2)*(J_scalar+(err[i])**2)
    J[ep]=J_scalar
    
    clear_output(wait=True)
    
    plt.scatter(x,fx, color= 'red', label = 'Sin(2 pi x )')
    plt.plot(x2, y2, 'b-')
    plt.title("Data set vs learned model")
    plt.xlabel("x")
    plt.ylabel("Polinomical function")
    plt.show()
```
